```python
import pandas as pd
```

```python
df=pd.read_csv('/content/servicos_setembro_pr_curitiba.txt')
```

```python
df=df[df['matricula_tecnico']!='EXTERNO']
```

```python
# Converta a coluna 'dt_fim_execucao' para o formato de data e hora
df['dt_fim_execucao'] = pd.to_datetime(df['dt_fim_execucao'], format='%Y-%m-%d %H:%M:%S.%f %Z', utc=True, errors='coerce').copy()

# Ordene o DataFrame pela coluna 'dt_fim_execucao' em ordem crescente
df = df.sort_values(by='dt_fim_execucao', ascending=True)
```

```python
df
```

```python
specific_date = pd.to_datetime('2023-09-02', utc=True)
df = df[df['dt_fim_execucao'].dt.date == specific_date.date()]
```
